```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('supermarket.csv')
df
```

```python
# Setting option to show all the columns
pd.set_option('display.max_columns', None)

df
```

```python
# Replacing all the ? with NaN
df.replace('?', np.nan, inplace=True)
df
```

```python
# Applying Apriori Principle to find the most frequent itemsets and association rules
one_rules = df.apply(lambda col: col.eq('t').sum(), axis=0).nlargest(10)
rules = one_rules.to_dict()
rules
```

```python
def support(itemset, otherset):
    merged = itemset + otherset
    return df.groupby(merged).size()
```

```python
# Applying Apriori Algorithm for two items dataset using the items in above rules dictionary
items = list(rules.keys())

for i in range(len(items)):
    for j in range(i+1, len(items)):
        item1 = [items[i]]
        item2 = [items[j]]

        # Calculating support for item1 and item2
        support_item1 = support(item1, item2)
        support_item2 = support(item2, item1)

        print(support_item1)
```
